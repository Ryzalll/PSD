# outlier-detection

```python
import pandas as pd
import sys
print(sys.version)
```

```python
from pycaret.anomaly import *
df = pd.read_csv('../../Dataset/iris-full.csv')

exp_name = setup(data = df)
abod = create_model('abod')
plot_model(abod, plot = 'tsne')
```

## Implementasi Model ke dataset

```python
# Assign model -> hasilkan dataframe dengan kolom anomaly
df_anomaly = assign_model(abod)
```

## Menampilkan Data dengan skor outlier tertinggi

```python
df_sorted = df_anomaly.sort_values(by="Anomaly_Score", ascending=False).reset_index(drop=True)
df_sorted.head()
```

## Menghapus Data dengan 2 skor outlier tertinggi

```python
df_removed = df_sorted.iloc[2:].reset_index(drop=True)
dataset = df_removed.sort_values(by="id").reset_index(drop=True)
dataset.head()
```
